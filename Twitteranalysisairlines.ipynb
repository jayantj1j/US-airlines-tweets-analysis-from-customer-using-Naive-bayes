```python
import nltk

from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def simple(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
        
```

```python

def clean_review(dd):
    out=[]
    for e in dd:
        if e.lower() in stop or  e.isalnum==True or e.isdigit()==True:
            continue
        else:    
            pos=pos_tag([e])
            clean=lem.lemmatize(e,pos=simple(pos[0][1]))
            out.append(clean.lower())
    return out    
```

```python
dataformat=[(clean_review(document),category) for document,category in dataformat]
dataformat[0:5]
```

```python
allwords=[]
for doc in dataformat:
    allwords+=doc[0]
```

```python
import re

text = u'This dog \U0001f602 \U0001F64F'
print(text) # with emoji

emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  
        u"\U0001F300-\U0001F5FF"  
        u"\U0001F680-\U0001F6FF"  
        u"\U0001F1E0-\U0001F1FF" "]+", flags=re.UNICODE)
print(emoji_pattern.sub('', text))
```

```python
freq=(nltk.FreqDist(allwords)) #type is freq distribution object
common=freq.most_common(3000)
features=[i[0] for i in common]
```

```python
features
fin=[e for e in features if len(e)>2]
fin.remove('united')
fin.remove('usairways')
fin.remove('americanair')
fin.remove('jetblue')
fin.remove('southwestair')
fin
```

```python
def get_feature_dict(words):
    current_features={}
    words_set=set(words)
    for w in fin:
        current_features[w]=w in words_set
    return current_features    
        
```

```python
ttrainfinal=[(get_feature_dict(tt),cat)for tt,cat in dataformat]
```

```python
from nltk import NaiveBayesClassifier
classfier=NaiveBayesClassifier.train(ttrainfinal)
```
